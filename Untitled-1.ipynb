```python
#split the data
x=df.drop(["Diagnosis"],axis=1)
y=df["Diagnosis"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#select model
model1=DecisionTreeClassifier()
model1.fit(x_train,y_train)
```

```python
#test the model
y_pred=model1.predict(x_test)
print(y_pred)
```

```python
#cross validation 
k=KFold(n_splits=5)
cor=cross_val_score(model1,x,y,cv=5)
print("accuracy_score:",cor)
```
